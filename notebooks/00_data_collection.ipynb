# Financial Statement Fraud Detection  

This notebook performs the **data collection step for financial statements using the VNStock API**.

### Objectives
- Retrieve **financial statement data of listed companies**
- Remove companies **in the financial sector** (banks, securities firms, insurance companies, etc.)
- Create the initial dataset for the **Financial Statement Fraud Detection** task

### Data Source
- Library: `vnstock`
- Data provider: `VCI`
- Financial reports used:
  - Balance Sheet
  - Income Statement
  - Cash Flow Statement

## Load Stock Symbol List

The stock symbol list is loaded from the file: `all_symbols.xlsx`.


```python
symbols_file = PROJECT_ROOT / "data/all_symbols.xlsx"

symbols = load_symbols_from_excel(
    excel_path=symbols_file,
    top_n=500,
)

print("Number of filtered non-financial symbols:", len(symbols))
print(symbols[:20])
```

## Financial Data Collection

At this stage, the system will:

1. Iterate through each stock symbol
2. Download the following data:
   - Balance Sheet
   - Income Statement
   - Cash Flow Statement
3. Standardize the data
4. Create **financial ratios**
5. Generate the **Fraud label**

```python
# Chạy thử ít mã trước để kiểm tra
sample_symbols = symbols

raw_df, failed_symbols = collect_financial_dataset(
    symbols=sample_symbols,
    save_path=RAW_DATA_DIR / "raw_data.xlsx",
)

print("Collected shape:", raw_df.shape)
print("Failed symbols:", failed_symbols)
display(raw_df.head(20))
```
